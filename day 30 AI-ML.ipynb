```python
from sklearn.datasets import load_digits
digit=load_digits()
print(digit.DESCR)
```

```python
print(digit.feature_names)
```

```python
print(digit.target_names)
```

```python
import pandas as pd
df=pd.DataFrame(digit.data)
df.head()
```

```python
df.columns=digit.feature_names
df.head()
```

```python
digit.data.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(digit.data,digit.target,stratify=digit.target,train_size=0.7,random_state=42)

log_reg=LogisticRegression()
log_reg.fit(x_train,y_train)
y_pred=log_reg.predict(x_test)
```

```python
df_predicted=pd.DataFrame()
df_predicted['Actual']=y_test
df_predicted['Predicted']=y_pred
df_predicted.head()
```

```python
#The method without using any functions !
import numpy as np
mislabel=np.sum((y_test!=y_pred))
print("Total number of mislabelled data points from {} test samples is {}".format(len(y_test),mislabel))
```

```python
print("Accuracy= ",517/540)
```

```python
#calculating accuracy with the help of functions !
from sklearn.metrics import accuracy_score
print("Accuracy of the test set is: ",accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cmdf=pd.DataFrame(cm,index=digit.target_names,columns=digit.target_names)
cmdf
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
target_probabilities=log_reg.predict_proba(x_test)[:,1]
false_positive_rate, true_positive_rate, threshold=roc_curve(y_test,target_probabilities)

%matplotlib inline
import matplotlib.pyplot as plt

plt.title("ROC")
plt.plot(false_positive_rate,true_positive_rate)
plt.plot([0,1], ls="--")
plt.plot([0, 0], [1, 0] , c=".7") 
plt.plot([1, 1] , c=".7")
plt.ylabel("True Positive Rate")
plt.xlabel("False Positive Rate")
plt.show()
```
