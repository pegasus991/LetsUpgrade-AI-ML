```python
fd =df.groupby(['Transaction','Item']).size().reset_index(name='count')
fd
```

```python
basket = (fd.groupby(['Transaction', 'Item'])['count'].sum().unstack().reset_index().fillna(0).set_index('Transaction'))
basket
```

```python
basket_sets = basket.applymap(lambda x:1 if x>0 else 0)
basket_sets
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="lift")
rules
```
