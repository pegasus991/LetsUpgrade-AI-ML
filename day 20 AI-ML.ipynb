## Q1. (approach 1)

```python
import pandas as pd
data = pd.DataFrame({'Pat':[5, 4, 4, 3, 9, 4],
'Jack':[4, 8, 7, 5, 1, 5],
'Alex':[9, 9, 8, 10, 4, 10]})
data
```

```python
data_new = data.melt(id_vars=None , value_vars=['Pat','Jack','Alex'])
data_new
```

```python
data_new.groupby('variable').mean()
```

```python
import statsmodels.api as sm
from  statsmodels.formula.api import ols
```

```python
ln = ols('value ~ variable', data=data_new).fit()
table=sm.stats.anova_lm(ln)
table
```

### Approach 2

```python
import pandas as pd

Pat = [5, 4, 4, 3, 9, 4]
Jack = [4, 8, 7, 5, 1, 5]
Alex = [9, 9, 8, 10, 4, 10]
all_scores = Pat + Jack + Alex
name = (['Pat']*len(Pat))+(['Jack']*len(Jack))+(['Alex']*len(Alex))

data1 = pd.DataFrame({'name_col':name , 'score':all_scores})
data1
```

```python
data1.groupby('name_col').mean()
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
lns=ols('score ~ name_col',data=data1).fit()
table1=sm.stats.anova_lm(lns)
table1
```

From both the approaches we get the same answer . Now looking at the P value......since the P value 0.029668 is less than 0.05 hence we will reject the null hypothesis . This means that the MEANS of all the population are not equal.....hence one or more groups have different mean
